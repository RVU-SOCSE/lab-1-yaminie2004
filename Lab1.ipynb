```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Dataset
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([52, 55, 61, 70, 82])

# Model A: Linear Regression
linear_model = LinearRegression()
linear_model.fit(x, y)

y_pred_linear = linear_model.predict(x)

print("MODEL A: SIMPLE LINEAR REGRESSION")
print("Equation: y =", linear_model.intercept_, "+", linear_model.coef_[0], "x")
print("MSE (Linear):", mean_squared_error(y, y_pred_linear))
print()

# Model B: Polynomial Regression (Degree 4)
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(x)

poly_model = LinearRegression()
poly_model.fit(x_poly, y)

y_pred_poly = poly_model.predict(x_poly)

print("MODEL B: POLYNOMIAL REGRESSION (Degree 4)")
print("MSE (Polynomial):", mean_squared_error(y, y_pred_poly))
print()

# Prediction for x = 6
x_new = np.array([[6]])
print("Prediction when x = 6")
print("Linear Model Prediction:", linear_model.predict(x_new))
print("Polynomial Model Prediction:", poly_model.predict(poly.transform(x_new)))

# Plot
plt.scatter(x, y)
plt.plot(x, y_pred_linear)
plt.plot(x, y_pred_poly)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Biasâ€“Variance Trade-off")
plt.show()
```
